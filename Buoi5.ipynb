```python
#1
def Max_Of_Two(a, b):
    return a if a > b else b
```

```python
#2
def swap(a, b):
    return b, a
```

```python
#3
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
```

```python
#4
def is_perfect(n):
    if n < 2:
        return False
    s = 0
    for i in range(1, n):
        if n % i == 0:
            s += i
    return s == n
```

```python
#5
def find_first(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1
```

```python
#6
def factorial(n):
    f = 1
    for i in range(1, n+1):
        f *= i
    return f
```

```python
#7
def calculator(num1, num2, n):
    if n == '+':
        return round(num1 + num2, 2)
    elif n == '-':
        return round(num1 - num2, 2)
    elif n == '*':
        return round(num1 * num2, 2)
    elif n == '/':
        if num2 == 0:
            return "Error"
        return round(num1 / num2, 2)
```

```python
#9
def sum_of_digits(n):
    s = 0
    n = abs(n)
    while n > 0:
        s += n % 10
        n //= 10
    return s
```
