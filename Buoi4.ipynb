```python
# Bài 6
def kt_nt(x):
    if x < 2:
        return False
    for i in range(2, int(x**0.5)+1):
        if x % i == 0:
            return False
    return True

a, b = map(int, input().split())
tong = 0
for i in range(a, b+1):
    if kt_nt(i):
        tong += i
print(tong)
```

```python
# Bài 7
n = int(input())
uoc = 2
ket_qua = 1

while n > 1:
    if n % uoc == 0:
        ket_qua = uoc
        while n % uoc == 0:
            n //= uoc
    uoc += 1 if uoc == 2 else 2

print(ket_qua)
```

```python
# Bài 8
def dao_so(x):
    dao = 0
    while x > 0:
        dao = dao * 10 + x % 10
        x //= 10
    return dao

def la_palindrome(x):
    return x == dao_so(x)

n = int(input())
buoc = 0

while not la_palindrome(n):
    n = n + dao_so(n)
    buoc += 1

print(buoc, n)
```

```python
# Bài 9
def chu_so_khac_nhau(x):
    cs = set()
    while x > 0:
        c = x % 10
        if c in cs:
            return False
        cs.add(c)
        x //= 10
    return True

n = int(input())
kq = []
i = 1
while i * i <= n:
    b = i * i
    if chu_so_khac_nhau(b):
        kq.append(b)
    i += 1

if kq:
    print(' '.join(map(str, kq)))
else:
    print("(no number)")
```

```python
# Bài 10
def do_dai_collatz(x):
    d = 1
    while x != 1:
        if x % 2 == 0:
            x //= 2
        else:
            x = 3 * x + 1
        d += 1
    return d

n = int(input())
max_len = 0
so_max = 1

for i in range(1, n+1):
    d = do_dai_collatz(i)
    if d > max_len:
        max_len = d
        so_max = i

print(so_max, max_len)
```

```python
# Bài 30
chuoi = input()
hoa = 0
thuong = 0
so = 0
for ky_tu in chuoi:
    if 'A' <= ky_tu <= 'Z':
        hoa += 1
    elif 'a' <= ky_tu <= 'z':
        thuong += 1
    elif '0' <= ky_tu <= '9':
        so += 1
print(hoa, thuong, so)
```

```python
# Bài 32
mk = input()
dac_biet = "!@#$%^&*()-_=+[]{}|;:,.<>?/"
co_hoa = False
co_thuong = False
co_so = False
co_db = False

if len(mk) <= 6:
    print("False")
else:
    for kt in mk:
        if 'A' <= kt <= 'Z':
            co_hoa = True
        elif 'a' <= kt <= 'z':
            co_thuong = True
        elif '0' <= kt <= '9':
            co_so = True
        elif kt in dac_biet:
            co_db = True

    print(co_hoa and co_thuong and co_so and co_db)
```
