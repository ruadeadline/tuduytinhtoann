```python
#W2A3
a = int(input('Nhập số nguyên a: '))
b = int(input('Nhập số nguyên b: '))
print('Tổng:', a + b)
print('Hiệu:', a - b)
print('Tích:', a * b)
if b != 0:
    print('Thương:', round(a / b, 2))
    print('Phần nguyên:', a // b)
    print('Phần dư:', a % b)
else:
    print('Không thể chia cho 0')
```
