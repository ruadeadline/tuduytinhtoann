```python
def binary_search(arr, left, right, target):
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# Nhập liệu
try:
    arr = list(map(int, input().split()))
    target = int(input())
    result = binary_search(arr, 0, len(arr) - 1, target)
    print(f"{result}")
except ValueError:
    print("Vui lòng nhập số nguyên hợp lệ.")
```

```python
def count_occurrences(arr, x):
    count = 0
    for num in arr:
        if num == x:
            count += 1
    return count

# Nhập liệu
arr = list(map(int, input().split()))
x = int(input())
print(f"{count_occurrences(arr, x)}")
```

```python
#W7A3
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Nhập liệu
data = list(map(int, input().split()))
print(f"{bubble_sort(data)}")
```

```python
# W7A4
def find_most_frequent(arr):

    counts = {}
    max_count = 0
    first_val = arr[0]

    for x in arr:
        counts[x] = counts.get(x, 0) + 1
        if counts[x] > max_count:
            max_count = counts[x]
            first_val = x

    return f"{first_val} xuat hien nhieu nhat, som nhat, {max_count} lan"

# Nhập liệu
arr = list(map(int, input("Nhập danh sách: ").split()))
print(find_most_frequent(arr))
```

```python
#W7A5
def count_pairs_with_sum(arr, x):
    # Sử dụng từ điển (dictionary) để lưu tần suất xuất hiện của các số đã duyệt
    seen_numbers = {}
    count = 0

    for num in arr:
        target = x - num
        # Nếu số cần tìm (target) đã tồn tại trong các số đã duyệt
        if target in seen_numbers:
        # Cộng số lần xuất hiện của target vào biến đếm
            count += seen_numbers[target]

        # Cập nhật số lần xuất hiện của số hiện tại vào từ điển
        seen_numbers[num] = seen_numbers.get(num, 0) + 1

    return count
try:
    input_str = input()
    arr = list(map(int,input_str.split()))
    x = int(input())
    result = count_pairs_with_sum(arr, x)
    print(f"Số lượng cặp (a, b) có tổng bằng {x} là: {result}")

except ValueError:
    print("Vui lòng chỉ nhập các số nguyên hợp lệ!")
```

```python
#W7A7
def find_insertion_point(tails, x):
    """
    Tìm vị trí đầu tiên trong mảng 'tails' có giá trị >= x.
    """
    low = 0
    high = len(tails)

    while low < high:
        mid = (low + high) // 2
        if tails[mid] < x:
            low = mid + 1
        else:
            high = mid
    return low

def longest_increasing_subsequence(nums):
    if not nums:
        return 0

    # tails[i] lưu số hạng nhỏ nhất có thể kết thúc một dãy con tăng độ dài i+1
    tails = []

    for x in nums:
        idx = find_insertion_point(tails, x)

        # Nếu idx bằng độ dài hiện tại, x lớn hơn tất cả -> thêm vào cuối
        if idx == len(tails):
            tails.append(x)
        else:
            # Nếu không, cập nhật phần tử tại idx bằng x để tối ưu dãy con
            tails[idx] = x

    return len(tails)
data = input()
nums = [int(item) for item in data.split()]
result = longest_increasing_subsequence(nums)
print(f"{result}")
```

```python
ef parse_list(s):
    # Hàm thủ công để chuyển chuỗi "[[1,4],[2,4]]" thành list [[1,4],[2,4]]
    # Xóa bỏ các ký tự không cần thiết
    s = s.strip()
    if s.startswith('[['): # Xử lý danh sách lồng nhau
        s = s[2:-2]
        parts = s.split('], [')
        result = []
        for p in parts:
            result.append([int(x) for x in p.split(',')])
        return result
    else: # Xử lý danh sách đơn [2,3,4]
        s = s[1:-1]
        return [int(x) for x in s.split(',')]

def solve():
        line1 = input()
        line2 = input()

        intervals = parse_list(line1)
        queries = parse_list(line2)

        # Lưu kết quả cho từng truy vấn theo đúng thứ tự ban đầu
        results = []

        for q in queries:
            min_len = -1

            for l, r in intervals:
                # Kiểm tra nếu khoảng thời gian bao phủ điểm q
                if l <= q <= r:
                    length = r - l + 1
                    # Cập nhật độ dài nhỏ nhất tìm được
                    if min_len == -1 or length < min_len:
                        min_len = length

            results.append(min_len)

        print("Kết quả:", results)

if __name__ == "__main__":
    solve()
```
