```python
import math

class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
        self.pi = 3.14

    def surface_area(self):
        # Diện tích toàn phần = 2*pi*r*h + 2*pi*r^2
        return 2 * self.pi * self.radius * self.height + 2 * self.pi * (self.radius ** 2)

    def volume(self):
        # Thể tích = pi*r^2*h
        return self.pi * (self.radius ** 2) * self.height

# Nhập dữ liệu
    r, h = map(float, input().split())
    cyl = Cylinder(r, h)
    print(f"{cyl.surface_area():.2f} {cyl.volume():.2f}")
```

```python
class Rectangle:
    def __init__(self, w, h):
        self.w = float(w)
        self.h = float(h)

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return (self.w + self.h) * 2

    def scale(self, k):
        self.w *= k
        self.h *= k

# Đọc input: w h k
data = input().split()
if len(data) == 3:
    r = Rectangle(data[0], data[1])
    r.scale(float(data[2]))
    print(f"{r.area():.2f} {r.perimeter():.2f}")
```

```python
import math

class Fraction:
    def __init__(self, num, den):
        if den == 0: raise ValueError
        common = math.gcd(num, den)
        self.num = num // common
        self.den = den // common
        if self.den < 0:
            self.num = -self.num
            self.den = -self.den

    def __str__(self):
        return f"{self.num}/{self.den}"
```

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, x):
        self.balance += float(x)

    def withdraw(self, x):
        amount = float(x)
        if amount <= self.balance:
            self.balance -= amount

    def __str__(self):
        return f"{self.balance:.2f}"

# Xử lý theo số lệnh n
line1 = input().split()
owner, initial = line1[0], line1[1]
acc = BankAccount(owner, initial)
n = int(input())
for _ in range(n):
    cmd = input().split()
    if cmd[0] == "DEPOSIT": acc.deposit(cmd[1])
    else: acc.withdraw(cmd[1])
print(acc)
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, subject, score):
        self.scores.append(float(score))

    def avg(self):
        return sum(self.scores) / len(self.scores) if self.scores else 0

    def rank(self):
        a = self.avg()
        if a >= 8: return "Excellent"
        if a >= 6.5: return "Good"
        if a >= 5: return "Average"
        return "Poor"

# Input: Tên, Số môn, và danh sách môn:điểm
name = input().strip()
n = int(input())
stu = Student(name)
for _ in range(n):
    _, score = input().split()
    stu.add_score(_, score)
print(f"{stu.name} {stu.avg():.2f} {stu.rank()}")
```

```python
class ComplexNumber:
    def __init__(self, real, imag):
        self.real = float(real)
        self.imag = float(imag)

    def multiply(self, other):
        r = self.real * other.real - self.imag * other.imag
        i = self.real * other.imag + self.imag * other.real
        return ComplexNumber(r, i)

# Nhập 2 dòng, mỗi dòng là một số phức
c1_data = input().split()
c2_data = input().split()
c1 = ComplexNumber(c1_data[0], c1_data[1])
c2 = ComplexNumber(c2_data[0], c2_data[1])
res = c1.multiply(c2)
print(f"{res.real:.0f} {res.imag:.0f}")
```

```python
class Triangle:
    def __init__(self, a, b, c):
        self.a, self.b, self.c = a, b, c

    def getArea(self):
        a, b, c = self.a, self.b, self.c
        if a+b > c and a+c > b and b+c > a:
            s = (a + b + c) / 2
            area = (s * (s - a) * (s - b) * (s - c)) ** 0.5
            return area
        return -1

# Nhập 3 cạnh
    sides = list(map(int, input().strip()))
    tri = Triangle(sides[0], sides[1], sides[2])
    res = tri.getArea()
    print(f"{res:.2f}" if res != -1 else "invalid")
```

```python
class Calculator:
    def __init__(self, num1=0, num2=0):
        # Khởi tạo lớp với hai thuộc tính là hai số cần tính
        self.num1 = float(num1)
        self.num2 = float(num2)

    def set_numbers(self, num1, num2):
        # Cập nhật 2 số mới nếu người dùng muốn tính toán trên 2 số khác
        self.num1 = float(num1)
        self.num2 = float(num2)

    def add(self): return self.num1 + self.num2 # Phép cộng
    def subtract(self): return self.num1 - self.num2 # Phép trừ
    def product(self): return self.num1 * self.num2 # Phép nhân

    def divide(self):
        # Phép chia, kiểm tra chia cho 0
        return self.num1 / self.num2 if self.num2 != 0 else "Lỗi chia cho 0"

    def power(self): return self.num1 ** self.num2 # Hàm mũ
    def mod(self): return self.num1 % self.num2 # Lấy phần dư

def menu_calculator():
    calc = Calculator()
    while True:
        print("\n--- MENU MÁY TÍNH ---")
        print("1. Nhập/Cập nhật số")
        print("2. Cộng | 3. Trừ | 4. Nhân | 5. Chia | 6. Mũ | 7. Chia dư")
        choice = input("Chọn thao tác (1-7): ")

        if choice == '1':
            n1, n2 = map(float, input("Nhập hai số cách nhau bởi dấu cách: ").split())
            calc.set_numbers(n1, n2)
        elif choice == '2': print("Kết quả:", calc.add())
        elif choice == '3': print("Kết quả:", calc.subtract())
        elif choice == '4': print("Kết quả:", calc.product())
        elif choice == '5': print("Kết quả:", calc.divide())
        elif choice == '6': print("Kết quả:", calc.power())
        elif choice == '7': print("Kết quả:", calc.mod())

        exit_choice = input("Bạn có muốn thoát? (yes/no): ").lower()
        if exit_choice in ["yes"]: # Dừng khi gõ "yes", "YES" hoặc "Yes"
            break
```

```python
class ShoppingCart:
    def __init__(self):
        # 1. Khởi tạo với danh sách rỗng
        self.items = []

    def add_product(self, name, price):
        # 2. Thêm vào giỏ hàng một sản phẩm mới
        self.items.append({"name": name, "price": float(price)})
        print(f"Đã thêm {name} vào giỏ.")

    def remove_product(self, name):
        # 3. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên)
        original_count = len(self.items)
        self.items = [item for item in self.items if item['name'] != name]
        if len(self.items) < original_count:
            print(f"Đã xóa sản phẩm: {name}")
        else:
            print("Không tìm thấy sản phẩm.")

    def is_empty(self):
        # 4. Kiểm tra xem giỏ hàng có rỗng không
        return len(self.items) == 0

    def total_price(self):
        # 5. Tính tổng tiền giỏ hàng
        return sum(item['price'] for item in self.items)

    def display_all(self):
        # 6. Hiển thị toàn bộ sản phẩm trong giỏ
        if self.is_empty():
            print("Giỏ hàng đang trống.")
        else:
            print("Danh sách sản phẩm:")
            for item in self.items:
                print(f"- {item['name']}: {item['price']}")

    def clear_cart(self):
        # 7. Xóa toàn bộ giỏ hàng
        self.items = []
        print("Đã dọn sạch giỏ hàng.")

def menu_shopping():
    cart = ShoppingCart()
    while True:
        print("\n--- MENU GIỎ HÀNG ---")
        print("2. Thêm | 3. Xóa | 4. Kiểm tra rỗng | 5. Tính tổng | 6. Hiển thị | 7. Xóa hết")
        choice = input("Chọn (2-7): ")

        if choice == '2':
            name = input("Tên SP: "); price = input("Giá: ")
            cart.add_product(name, price)
        elif choice == '3':
            name = input("Tên SP cần xóa: ")
            cart.remove_product(name)
        elif choice == '4':
            print("Giỏ rỗng?" , cart.is_empty())
        elif choice == '5':
            print("Tổng tiền:", cart.total_price())
        elif choice == '6':
            cart.display_all()
        elif choice == '7':
            cart.clear_cart()

        exit_choice = input("Có muốn thoát khỏi chương trình không? (yes/no): ").lower()
        if exit_choice == "yes": # Chương trình dừng khi nhập "yes"
            break
```
