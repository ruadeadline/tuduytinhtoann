```python
try:
    a = int(input())
    b = int(input())
    result = a / b
    print(f"{result:.2f}")
except ZeroDivisionError:
    print("Loi chia khong")
except ValueError:
    print("Vui long nhap so nguyen")
```

```python
try:
    my_list = [int(x) for x in input().split()]
    n = len(my_list)
    print(my_list[n]) # Gây lỗi vì chỉ số tối đa là n-1
except IndexError:
    print("Truy cap chi so sai")
```

```python
def is_happy(n):
    seen = set()
    while n != 1 and n not in seen:
        seen.add(n)
        n = sum(int(digit)**2 for digit in str(n))
    return n == 1

try:
    a = input()
    if is_happy(int(a)):
        print("YES")
    else:
        print("NO")
except ValueError:
    print("Loi nhap lieu")
```

```python
data = ["10", "abc", "20", "3.5", "50"]
result = []
for item in data:
    try:
        result.append(int(item))
    except ValueError:
        print(f"Canh bao: '{item}' khong the chuyen sang int")
print("Danh sach sau khi loc:", result)
```

```python
import math
try:
    x = float(input())
    if x < 0:
        raise ValueError("So am")
    print(f"{math.sqrt(x):.2f}")
except ValueError as e:
    print(e)
```
