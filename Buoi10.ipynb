```python
#1
def merge_sorted_lists(list1, list2):
    result = []
    i, j = 0, 0
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            result.append(list1[i])
            i += 1
        else:
            result.append(list2[j])
            j += 1
    # Thêm phần còn lại của danh sách chưa hết
    result.extend(list1[i:])
    result.extend(list2[j:])
    return result
```

```python
def is_sublist(list1, list2):
    n, m = len(list1), len(list2)
    for i in range(m - n + 1):
        if list2[i : i + n] == list1:
            return True
    return False
```

```python
def find_kth_largest(matrix, k):
    n = len(matrix)
    m = len(matrix[0])
    # Chuyển bài toán tìm số lớn thứ k thành tìm số nhỏ thứ target
    target = n * m - k + 1

    low, high = matrix[0][0], matrix[n-1][m-1]
    while low < high:
        mid = (low + high) // 2
        count = 0
        j = m - 1
        for i in range(n):
            while j >= 0 and matrix[i][j] > mid:
                j -= 1
            count += (j + 1)
        if count < target:
            low = mid + 1
        else:
            high = mid
    return low
```

```python
def smallest_number_after_swap(s):
    num = list(s)
    n = len(num)
    for i in range(n - 1):
        min_idx = i
        # Tìm chữ số nhỏ nhất ở phía sau
        for j in range(n - 1, i, -1):
            if num[j] < num[min_idx]:
                # Tránh hoán đổi số 0 lên đầu
                if i == 0 and num[j] == '0':
                    continue
                if min_idx == i or num[j] < num[min_idx]:
                    min_idx = j

        if min_idx != i:
            num[i], num[min_idx] = num[min_idx], num[i]
            break
    return "".join(num)
```

```python
def can_transform(a, b):
    n, m = len(a), len(b)
    dp = [[False] * (m + 1) for _ in range(n + 1)]
    dp[0][0] = True

    for i in range(n):
        for j in range(m + 1):
            if not dp[i][j]: continue
            if a[i].islower():
                dp[i+1][j] = True
            if j < m and a[i].upper() == b[j]:
                dp[i+1][j+1] = True

    return "YES" if dp[n][m] else "NO"
```

```python
def next_permutation(arr):
    n = len(arr)
    i = n - 2
    while i >= 0 and arr[i] >= arr[i+1]:
        i -= 1
    if i >= 0:
        j = n - 1
        while arr[j] <= arr[i]:
            j -= 1
        arr[i], arr[j] = arr[j], arr[i]
    arr[i+1:] = reversed(arr[i+1:])
    return arr
```

```python
def canPlaceFlowers(flowerbed, k):
    count = 0
    f = [0] + flowerbed + [0] # Thêm đệm hai đầu để xử lý biên
    for i in range(1, len(f) - 1):
        if f[i-1] == 0 and f[i] == 0 and f[i+1] == 0:
            f[i] = 1
            count += 1
    return count >= k
```

```python
def is_path_crossing(moves):
    x, y = 0, 0
    visited = {(0, 0)}
    for move in moves:
        if move == 'U': y += 1
        elif move == 'D': y -= 1
        elif move == 'R': x += 1
        elif move == 'L': x -= 1

        if (x, y) in visited:
            return True
        visited.add((x, y))
    return False
```
