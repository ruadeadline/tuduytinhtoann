```python
list1=list(map(int,input().split()))
k=int(input())
kq1=[]
while len(list1)>0:
  x=list1[0]
  if k-x not in list1:
    list1.remove(x)
  else:
    if x==k-x and list1.count(x) == 1:
      list1.remove(x)
    elif x==k-x:
      list1.remove(x)
      #list1.remove(x)
      id=(x,x)
      if id not in kq1:
        kq1.append((x,x))
    else:
      list1.remove(x)
      #list1.remove(k-x)
      id=(x,k-x)
      if id not in kq1:
        kq1.append((x,k-x))
kq1.sort()
print(kq1)

```
